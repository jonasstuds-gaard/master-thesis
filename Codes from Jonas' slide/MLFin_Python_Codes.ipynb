```python
# =========================
# Loops: for and while
# =========================
# for-loop: print each price
prices = [100, 102, 101]
for p in prices:
    print(p)
# Output: 100 102 101

# while-loop: accumulate until threshold
total = 0
i = 0
while total < 300 and i < len(prices):
    total += prices[i]
    i += 1
print("Accumulated:", total)
# Output: Accumulated: 303

# break / continue in a loop
for p in prices:
    if p == 102:
        continue    # skip 102
    if p > 101:
        break       # stop on price > 101
    print(p)
```

```python
# =========================
# Functions: Definition & Usage
# =========================
def compute_return(p0, p1):
    """
    Compute simple return: (p1 - p0) / p0
    """
    return (p1 - p0) / p0

# Use the function
r = compute_return(100, 110)
print("Return:", r)
# Output: Return: 0.1

# Vectorized example via loop
prices = [100, 105, 110]
returns = []
for i in range(len(prices)-1):
    returns.append(compute_return(prices[i], prices[i+1]))
print(returns)
# Output: [0.05, 0.047619047619047616]
```

```python
# =========================
# Function Arguments: Positional, Keyword & Defaults
# =========================
def compound_return(p0, p1, periods=1):
    """
    Compute compound return over 'periods' intervals
    """
    return (p1 / p0)**(1/periods) - 1

# Positional
print(compound_return(100, 121, 2))
# Output: 0.1

# Keyword (order doesn't matter)
print(compound_return(p1=121, p0=100, periods=2))
# Output: 0.1

# Default periods=1
print(compound_return(100, 110))
# Output: 0.1
```

```python
# =========================
# NumPy: Indexing & Slicing — Examples
# =========================
# 1D slicing examples
a = np.array([1,2,3,4,5])
print(a[1:4])      # [2 3 4]
print(a[::2])      # [1 3 5]
print(a[-2:])      # [4 5]

# 2D array of returns
r = np.array([[0.01,  0.02,  0.015],
              [-0.005, 0.01,  0.005]])
print(r[0, :])     # [0.01  0.02  0.015]
print(r[:, 1])     # [0.02  0.01]

# Boolean mask: positive returns
print(r[r > 0])
# [0.01  0.02  0.015  0.01   0.005]

# Fancy integer indexing
rows = np.array([0, 1])
cols = np.array([2, 0])
print(r[rows, cols])
# [ 0.015 -0.005]
```

```python
# =========================
# NumPy: Vectorized Arithmetic & Broadcasting
# =========================
# Sample price series
prices = np.array([100, 102, 101, 105])

# Element-wise operations
diffs       = prices[1:] - prices[:-1]
scaled      = prices * 1.01      # +1% to each price
price_plus = prices[1:] + diffs    # broadcast diffs to match size
print("Diffs:", diffs)
# Output: [ 2 -1  4]
print("Scaled:", scaled)
# Output: [101.   103.02 102.01 106.05]

# Broadcasting example: subtract mean
mean_price = prices.mean()
demeaned    = prices - mean_price
print("Demeaned:", demeaned)
# Output: Demeaned: [-2.  0. -1.  3.]
```

```python
# =========================
# NumPy: Ufuncs & Aggregations
# =========================
# Continuing from 'prices'
# Simple returns
returns   = diffs / prices[:-1]
log_ret   = np.log(prices[1:] / prices[:-1])

# Ufuncs
sqrt_ret  = np.sqrt(returns.clip(min=0))
exp_ret   = np.exp(returns)

# Aggregations
total_ret    = returns.sum()
mean_ret     = returns.mean()
cum_ret      = (1 + returns).cumprod() - 1

print("Returns:", returns)
# Output: [ 0.02   -0.00980392  0.03960396]
print("Log returns:", log_ret)
# Output: Log returns: [ 0.01980263 -0.0098523   0.03883983]
print("Total return:", total_ret)
# Output: 0.04980004
print("Cumulative return:", cum_ret)
# Output: Cumulative return: [0.02 0.01 0.05]
```

```python
# =========================
# NumPy: Linear Algebra (Part 1)
# =========================
# Vector dot product
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])
print(u.dot(v))
# Output: 32

# Matrix to vector multiplication
weights = np.array([0.6, 0.4])
r = np.array([[0.01,   0.02,   0.015],
              [-0.005, 0.01,   0.005]])
print(weights @ r)
# Output: [0.006  0.018  0.011] (this was the intentional error)

# Matrix to matrix multiplication
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
print(A @ B)
# Output:
# [[19 22]
#  [43 50]]

# Transpose of returns matrix
print(r.T)
# Output:
# [[ 0.01  -0.005]
#  [ 0.02   0.01 ]
#  [ 0.015  0.005]]
```

```python
# =========================
# NumPy: Linear Algebra (Part 2)
# =========================
# Covariance matrix of returns (assets * days)
cov = np.cov(r, rowvar=False)
print(cov)
# Output (example):
# [[2.50e-05 3.75e-05]
#  [3.75e-05 5.83e-05]]

# Solve linear system M * x = b
M = np.array([[2, 1],
              [1, 3]])
b = np.array([1, 2])
x = np.linalg.solve(M, b)
print(x)
# Output: [0.2 0.6]

# Matrix inverse
inv_M = np.linalg.inv(M)
print(inv_M)
# Output:
# [[ 0.6  -0.2]
#  [-0.2  0.4]]

# Eigenvalues and eigenvectors of covariance
eigvals, eigvecs = np.linalg.eig(cov)
print("Eigenvalues:", eigvals)
# Output: [8.27e-05 6.25e-07]  (example)
```

```python
# =========================
# Data Cleaning: Examples
# =========================
# Mixed-type example with missing values
df3 = pd.DataFrame({
  'Date':  ['2025-07-01', None, '2025-07-03'],
  'Close': ['100', 'nan', '105']
})

# 1) Drop rows where Date is missing
df3 = df3.dropna(subset=['Date'])

# 2) Convert 'Close' to numeric, coerce bad values, then forward-fill
df3['Close'] = pd.to_numeric(df3['Close'], errors='coerce')
df3['Close'] = df3['Close'].fillna(method='ffill', limit=1)

# 3) Parse 'Date' to datetime, coercing errors
df3['Date'] = pd.to_datetime(df3['Date'], errors='coerce', dayfirst=False)

# 4) Inspect final DataFrame
print(df3.info())
```

```python
# =========================
# Aggregation & Grouping — Part 1
# =========================
# Sample returns by ticker
df4 = pd.DataFrame({
    'Ticker': ['AAPL','MSFT','AAPL','MSFT'],
    'Return': [0.01, 0.02, 0.03, 0.015]
})

# 1) Mean per ticker
mean_ret = df4.groupby('Ticker')['Return'].mean()
print(mean_ret)

# 2) Sum per ticker
sum_ret = df4.groupby('Ticker')['Return'].sum()
print(sum_ret)

# 3) Count per ticker
count_ret = df4.groupby('Ticker')['Return'].count()
print(count_ret)
```

```python
# =========================
# Aggregation & Grouping — Part 2
# =========================
# 4) Multiple aggregations at once
agg_ret = df4.groupby('Ticker')['Return']\
             .agg(['mean','std','count'])
print(agg_ret)

# 5) Group by multiple keys
df5 = pd.DataFrame({
    'Sector': ['Tech','Tech','Fin','Fin'],
    'Ticker': ['AAPL','MSFT','JPM','BAC'],
    'Return': [0.01,0.015,0.005,0.007]
})
grp = df5.groupby(['Sector','Ticker'])['Return'].sum()
print(grp)

# 6) Pivot table (mean per sector / ticker)
pivot = pd.pivot_table(df5, values='Return', index='Sector', columns='Ticker')
print(pivot)
```

```python
# =========================
# Aggregation & Grouping — Part 3
# =========================
# 7) Crosstab alternative
ct = pd.crosstab(df5['Sector'], df5['Ticker'], values=df5['Return'], aggfunc='sum')
print(ct)

# 8) Merge mean returns into price table
df_prices = pd.DataFrame({
    'Ticker': ['AAPL','MSFT'],
    'Close':  [150, 300]
})
merged = df_prices.merge(mean_ret.rename('MeanReturn'), on='Ticker', how='left')
print(merged)

# 9) Concatenate additional returns
df_more = pd.DataFrame({'Ticker':['GOOG','TSLA'],
                        'Return':[0.005, 0.02]})
combined = pd.concat([df4, df_more], axis=0)
print(combined)
```
