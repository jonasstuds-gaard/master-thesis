**Comments**

Lasso produces a sparser model by setting many coefficients exactly to zero, leaving only a subset of strong predictors. Lasso reduce variance, but it do so at the cost of increasing bias—since coefficients are pulled toward zero. Even the strongest predictors are partially shrunk, which can distort true effect sizes and make direct interpretation of coefficients less meaningful.

This involves refitting a standard OLS model only on the variables selected by the shrinkage model. Post-Lasso typically improves interpretability, reduces bias, and can further improve predictive accuracy if the initial selection was effective.

```python
from xgboost import XGBRegressor
# --- XGBoost model ---
xgb = XGBRegressor(
    objective="reg:squarederror", # Default
    n_estimators=200,
    random_state=42,
    n_jobs=-1,
    bootstrap=True
)
xgb.fit(X_train_3, y_train_3)

# --- Predictions ---
y_pred_train = xgb.predict(X_train_3)
y_pred_test = xgb.predict(X_test_3)

# --- Evaluation ---
mse_train = mean_squared_error(y_train_3, y_pred_train)
mse_test = mean_squared_error(y_test_3, y_pred_test)
rmse_train = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)

print("XGBoost:")
print(f" Train RMSE: {rmse_train:.6f}")
print(f" Test RMSE: {rmse_test:.6f}\n")
```

**Comments**

It appears to overfit as is training RMSE is lower than the test.
