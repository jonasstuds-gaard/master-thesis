```python
"""
NumPy Math Functions and Conditional Logic Reference
----------------------------------------------------
A comprehensive Python reference file for:
1. Mathematical functions in NumPy
2. If / elif / else condition examples with arrays and dictionaries
"""

import numpy as np

# ===============================================================
# 1. BASIC ARITHMETIC OPERATIONS
# ===============================================================

a, b = np.array([1, 2, 3]), np.array([4, 5, 6])
sum_ = a + b              # addition
diff = a - b              # subtraction
prod = a * b              # multiplication
quot = a / b              # division
power = a ** 2            # exponentiation
floor_div = a // 2        # floor division
mod = a % 2               # modulus

# ===============================================================
# 2. POWER, ROOT, AND EXPONENTIAL FUNCTIONS
# ===============================================================

x = np.array([1, 2, 3])
np.power(x, 3)            # x^3
np.sqrt(x)                # √x
np.cbrt(x)                # cube root
np.exp(x)                 # e^x
np.expm1(x)               # e^x - 1 (accurate for small x)
np.exp2(x)                # 2^x

# ===============================================================
# 3. LOGARITHMIC FUNCTIONS
# ===============================================================

np.log(x)                 # natural log (ln x)
np.log10(x)               # base 10 log
np.log2(x)                # base 2 log
np.log1p(x)               # log(1 + x) for small x

# ===============================================================
# 4. TRIGONOMETRIC FUNCTIONS
# ===============================================================

angle = np.pi / 4
np.sin(angle)
np.cos(angle)
np.tan(angle)

# Inverse trig
np.arcsin(1)
np.arccos(0)
np.arctan(1)
np.arctan2(1, 1)

# Hyperbolic
np.sinh(x)
np.cosh(x)
np.tanh(x)

# Inverse hyperbolic
np.arcsinh(x)
np.arccosh(np.array([1, 2, 3]))
np.arctanh(np.array([0.1, 0.2, 0.3]))

# ===============================================================
# 5. ROUNDING AND REMAINDER FUNCTIONS
# ===============================================================

y = np.array([1.2, 2.5, 3.7, -1.5])
np.round(y)
np.floor(y)
np.ceil(y)
np.trunc(y)
np.rint(y)
np.fix(y)
np.remainder(y, 2)
np.mod(y, 2)

# ===============================================================
# 6. ABSOLUTE VALUE AND SIGN
# ===============================================================

np.abs(y)
np.sign(y)

# ===============================================================
# 7. SPECIAL MATHEMATICAL FUNCTIONS
# ===============================================================

np.degrees(np.pi)
np.radians(180)
np.hypot(3, 4)

# ===============================================================
# 8. COMPARISON AND LOGICAL OPERATIONS
# ===============================================================

x, y = np.array([1, 3, 5]), np.array([2, 2, 6])
np.maximum(x, y)
np.minimum(x, y)
np.clip(x, 2, 4)
np.where(x > 2, x, 0)
np.isfinite(x)
np.isnan(x)
np.isinf(x)

# ===============================================================
# 9. STATISTICAL AND AGGREGATION FUNCTIONS
# ===============================================================

data = np.array([1, 2, 3, 4, 5])
np.sum(data)
np.mean(data)
np.std(data)
np.var(data)
np.median(data)
np.prod(data)
np.cumsum(data)
np.cumprod(data)

# ===============================================================
# 10. CONSTANTS
# ===============================================================

np.pi
np.e
np.inf
np.nan

# ===============================================================
# 11. EXAMPLES OF BUILDING CUSTOM FUNCTIONS
# ===============================================================

def f(x):
    """Example: f(x) = e^(-x²) * sin(2πx)"""
    return np.exp(-x**2) * np.sin(2 * np.pi * x)

def g(x):
    """Example: g(x) = sqrt(x) if x ≥ 0 else log(1 - x)"""
    return np.where(x >= 0, np.sqrt(x), np.log1p(-x))

def h(x):
    """Example: h(x) = 3x² + 2x + 1"""
    return 3 * np.power(x, 2) + 2 * x + 1

# ===============================================================
# 12. IF / ELIF / ELSE CONDITIONAL LOGIC
# ===============================================================

# ---------- Basic scalar example ----------
value = 10
if value > 20:
    result = "Large"
elif value > 10:
    result = "Medium"
else:
    result = "Small"

# ---------- Using NumPy arrays (vectorized) ----------
arr = np.array([-2, -1, 0, 1, 2, 3])

# Option 1: np.where
labels = np.where(arr < 0, "negative", "non-negative")

# Option 2: Nested np.where (if / elif / else)
labels2 = np.where(arr < 0, "negative",
           np.where(arr == 0, "zero", "positive"))

# Option 3: Logical conditions with masks
mask_pos = arr > 0
mask_neg = arr < 0
mask_zero = arr == 0

arr_labels = np.empty(arr.shape, dtype=object)
arr_labels[mask_pos] = "positive"
arr_labels[mask_neg] = "negative"
arr_labels[mask_zero] = "zero"

# ---------- Using loops for conditional logic ----------
classified = []
for val in arr:
    if val < 0:
        classified.append("negative")
    elif val == 0:
        classified.append("zero")
    else:
        classified.append("positive")

classified = np.array(classified)

# ---------- Example with mathematical thresholds ----------
grades = np.array([88, 72, 95, 60, 83])
grade_labels = np.where(grades >= 90, "A",
                np.where(grades >= 80, "B",
                np.where(grades >= 70, "C",
                np.where(grades >= 60, "D", "F"))))

# ===============================================================
# 13. CONDITIONAL LOGIC WITH DICTIONARIES
# ===============================================================

# ---------- Example 1: Simple if/else on dictionary values ----------
student = {"name": "Jonas", "score": 85}

if student["score"] >= 90:
    student["grade"] = "A"
elif student["score"] >= 80:
    student["grade"] = "B"
elif student["score"] >= 70:
    student["grade"] = "C"
else:
    student["grade"] = "F"

# ---------- Example 2: Iterating over a dictionary ----------
students = {
    "Alice": 92,
    "Bob": 75,
    "Charlie": 64,
    "David": 58
}

grades_dict = {}
for name, score in students.items():
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    grades_dict[name] = grade

# ---------- Example 3: Applying NumPy conditions to dictionary values ----------
scores = np.array(list(students.values()))
names = np.array(list(students.keys()))

grade_labels = np.where(scores >= 90, "A",
                np.where(scores >= 80, "B",
                np.where(scores >= 70, "C",
                np.where(scores >= 60, "D", "F"))))

grades_np_dict = dict(zip(names, grade_labels))

# ===============================================================
# 14. ADVANCED CONDITIONAL USE CASES
# ===============================================================

# Example: Assign risk categories
returns = np.array([-0.05, 0.02, 0.10, -0.02, 0.07])
risk_cat = np.where(returns < -0.02, "High Risk",
            np.where(returns < 0.05, "Medium Risk", "Low Risk"))

# Example: Transform dictionary based on multiple conditions
portfolio = {
    "AssetA": {"return": 0.03, "volatility": 0.15},
    "AssetB": {"return": 0.08, "volatility": 0.25},
    "AssetC": {"return": -0.01, "volatility": 0.30},
}

for asset, metrics in portfolio.items():
    r, v = metrics["return"], metrics["volatility"]
    if r < 0:
        metrics["status"] = "Loss"
    elif v > 0.2 and r < 0.05:
        metrics["status"] = "Risky"
    elif r >= 0.05 and v <= 0.2:
        metrics["status"] = "Safe"
    else:
        metrics["status"] = "Neutral"

# ===============================================================
# USAGE EXAMPLE
# ===============================================================

if __name__ == "__main__":
    x = np.linspace(-2, 2, 5)
    print("f(x):", f(x))
    print("g(x):", g(x))
    print("h(x):", h(x))
    print("Array classification:", arr_labels)
    print("Grades:", grade_labels)
    print("Grades dict:", grades_dict)
    print("Portfolio status:", portfolio)
```
